```python
import numpy as np
import pandas as pd
import re
import json
import requests
from bs4 import BeautifulSoup as bs
import warnings
warnings.filterwarnings('ignore')
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/84.0.4147.105 Safari/537.36'}

page_list = [""]
for i in np.arange(2,39):
    page_str = str(i) + "_p/"
    page_list.append(page_str)

page_list_url = []
for i in np.arange(len(page_list)):
    trulia_MonroeCounty_url = "https://www.trulia.com/County/NY/Monroe_Real_Estate/" + page_list[i]
    page_list_url.append(trulia_MonroeCounty_url)
```

```python
url_common = "https://www.trulia.com"
detail_url = []
location = []
latitude = []
longitude = []
bedrooms = []
bathrooms = []
space = []
price = []

for i in np.arange(38):
    r = requests.get(page_list_url[i],headers=headers)
    soup = bs(r.content,'html.parser')
    for j in np.arange(30):
        detail_info = json.loads(soup.find_all(type="application/json")[0].string).get("props").get("searchData").get("homes")
        detail_url.append(url_common + detail_info[j].get("url"))
        location.append(detail_info[j].get("location").get("fullLocation"))
        latitude.append(detail_info[j].get("location").get("coordinates").get("latitude"))
        longitude.append(detail_info[j].get("location").get("coordinates").get("longitude"))
        # in case there is no data in bedroom
        if detail_info[j].get("bedrooms") is None:
            bedrooms.append("None")
        else:
            bedrooms.append(detail_info[j].get("bedrooms").get("formattedValue").replace("bd",""))
        if detail_info[j].get("bathrooms") is None:
            bathrooms.append("None")
        else:
            bathrooms.append(detail_info[j].get("bathrooms").get("formattedValue").replace("ba",""))
        if detail_info[j].get("floorSpace") is None:
            space.append("None")
        else:    
            space.append(detail_info[j].get("floorSpace").get("formattedDimension").replace(",","").replace(" sqft",""))
        price.append(detail_info[j].get("price").get("price"))
```

```python
feature = ["detail_url", "location", "latitude", 'longitude', 'bedrooms', 'bathrooms', 'space', 'price']
df = pd.DataFrame(columns=feature)
df["detail_url"] = detail_url
df["location"] = location
df["latitude"] = latitude
df["longitude"] = longitude
df["bedrooms"] = bedrooms
df["bathrooms"] = bathrooms
df["space"] = space
df["price"] = price
df
```
